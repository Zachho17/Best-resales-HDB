```python
plt.figure(figsize=(10,10))
graph = sns.boxplot(data[~data['flat_type'].isin(['1 ROOM','MULTI-GENERATION'])], x='flat_type', y='resale_price')
graph.axhline(data[~data['flat_type'].isin(['1 ROOM','MULTI-GENERATION'])]['resale_price'].mean())
plt.xlabel('flat type')
plt.ylabel('resale price')
plt.title('Boxplot between flag type and resale price')
plt.show()
```

Above attempted to plot the boxplot that exclude 1-room HDB and Multi-generation HDB, since they seem to have relatively normal distribution.  
Then I added the average resale price of the remaining 5 flat type, hoping to further identify the reason(s) for these flats to be above average.  
However, in the above boxplot, it looks like using average price across is not sufficient.

```python
for type in types:
    upper = data[data['flat_type'] == type]['resale_price'].quantile(0.75)
    lower = data[data['flat_type'] == type]['resale_price'].quantile(0.25)
    iqr = upper - lower
    outliner = upper + iqr*1.5
    print(f'{type} type flat that above {outliner} is considered outliers')
```

From the above, we can isolate a new group of flats that is considered outliers in the respective flat types.  
From there, we would attempt to identify what makes them more expensive compared to the majority of the flats.

### by outliers

```python
data_outliers = data[
                    ((data['flat_type'] == '1 ROOM') & (data['resale_price'] > 297000)) |
                    ((data['flat_type'] == '2 ROOM') & (data['resale_price'] > 329000)) |
                    ((data['flat_type'] == '3 ROOM') & (data['resale_price'] > 463000)) |
                    ((data['flat_type'] == '4 ROOM') & (data['resale_price'] > 641000)) |
                    ((data['flat_type'] == '5 ROOM') & (data['resale_price'] > 803000)) |
                    ((data['flat_type'] == 'MULTI-GENERATION') & (data['resale_price'] > 960000)) |
                    ((data['flat_type'] == 'EXECUTIVE') & (data['resale_price'] > 885000))
                    ]
print(f'outliers data set has a shape of {data_outliers.shape}')
print(f'about {round((data_outliers.shape[0]/data.shape[0])*100, 2)}% of the total dataset')
```

### by planning area

```python
data_byarea_mid = data.groupby('planning_area').agg({'resale_price':'median'})
data_byarea_mid = data_byarea_mid.sort_values('resale_price', ascending=True)

data_byarea_mean = data.groupby('planning_area').agg({'resale_price':'mean'})
data_byarea_mean = data_byarea_mean.sort_values('resale_price', ascending=True)

fig = plt.figure(figsize = (20, 7))
fig , (ax1, ax2) = plt.subplots(1,2)
fig.tight_layout(w_pad=13)
                 
ax1.barh(data_byarea_mid.index.values, data_byarea_mid['resale_price'])
ax1.set_title('median price by area')
ax1.tick_params(labelsize=7)

ax2.barh(data_byarea_mean.index.values, data_byarea_mean['resale_price'])
ax2.set_title('mean price by area')
ax2.tick_params(labelsize=7)
```

From the above, it looks like certain areas such as Tanglin, Bukit Timah and Outram tend to have both higher median and mean price  
Meanwhile Changi, Yishun, Bukit Batok, Woodlands and Ang Mo Kio look to offer the cheapest resale flat.  
It is worth noting though, across data is across 2012 to 2021, so the median might not reflect the recent price information  
Moreover, with mean, area with less number of HDB flats might have relatively higher mean due to smaller denominator  

However, this does give us a clue on location could be a factor that impact HDB prices.
